# Distribuição agrupada por palavras

```python
import matplotlib.pyplot as plt
import numpy as np

def receberDict(x=1):
    if x == 1:
        x = input("Arquivo a ser lido: ")
    leitor = open("{}".format(x), "r", encoding="utf-8")
    texto = leitor.read()
    for letra in "(),":
        texto = texto.replace(letra, " ")
    texto = texto.split()

    dicionario = {}
    for i in range(0, len(texto), 2):
        chave = texto[i].strip("'")
        valor = int(texto[i + 1])
        dicionario[chave] = valor
    return dicionario

# função de massa de probabilidade padrão para a distribuição de zipf
def porcZipf(n, rank=1):
    if n > 0:
        total = 0
        for j in range(1, n+1):
            total += 1/j
        return 1/(total*rank)
    else:
        print("Calculo impossivel")

#Os parametros y e a não são necessário e servem para transformar na variacao Zipf-Mandelbrot
def distribuicao_zipf(n, y=1, a=0):
    # Gera os ranks e probabilidades de acordo com a lei de Zipf
    ranks = np.arange(1 + a, n + 1 + a)
    probabilities = 1 / (ranks ** y)

    # Normaliza as probabilidades para que somem 1
    probabilities /= np.sum(probabilities)
    return ranks, probabilities

def plot_zipf_distribution(n, y):
    ranks, probabilities = distribuicao_zipf(n, y)

    # Plota a distribuição de Zipf
    plt.plot(ranks, probabilities, color='blue', alpha=0.7)
    plt.xscale('log')
    plt.yscale('log')
    plt.title(f'Distribuição de Zipf {soma} palavras totais e {n} palavras unicas em {nomeLivro} fator y={y}')
    plt.xlabel('Rank')
    plt.ylabel('Probabilidade')

# Mude o caminho para fazer a distribuicao com outros livros
livro = "resultados/saida-casmurro.txt"

#recebe apenas o nome do livro sem o caminho
nomeLivro = livro.split("-")[1]

dicionario = receberDict(livro)
listaValores = list(dicionario.values())
listaChaves = list(dicionario.keys())

tamanho = len(listaValores)
soma = sum(listaValores)

# Usando enumerate para obter índices e valores simultaneamente
lista = [i + 1 for i, _ in enumerate(listaValores)]
listaValores = [val / soma for val in listaValores]

# Plota o gráfico de pontos vermelhos
plt.plot(lista, listaValores, 'ro')

n_palavras_diferentes = len(listaValores)
parametro_y = 1

# Chama a função para plotar a distribuição de Zipf no mesmo gráfico
plot_zipf_distribution(n_palavras_diferentes, parametro_y)

# Exibe o gráfico final
plt.show()
```

# Distribuição agrupada em pares de palavras

```python
import matplotlib.pyplot as plt
import numpy as np

def receberDictPar(x=1):
    if x == 1:
        x = input("Arquivo a ser lido: ")
    leitor = open("{}".format(x), "r", encoding="utf-8")
    texto = leitor.read()
    for letra in "(),":
        texto = texto.replace(letra, " ")
    texto = texto.split()

    dicionario = {}
    i = 0
    while i < len(texto):
        # Verifica se a chave é composta por duas palavras consecutivas
        chave = f"{texto[i]} {texto[i + 1].strip(',')}"
        valor_str = texto[i + 2]

        # Verifica se o valor é composto apenas por caracteres numéricos
        if valor_str.isdigit():
            valor = int(valor_str)
            dicionario[chave] = valor
        else:
            print(f"Valor inválido para a chave '{chave}': {valor_str}")

        i += 3  # Avança para o próximo conjunto de palavras e valor

    return dicionario

# função de massa de probabilidade padrão para a distribuição de zipf
def porcZipf(n, rank=1):
    if n > 0:
        total = 0
        for j in range(1, n+1):
            total += 1/j
        return 1/(total*rank)
    else:
        print("Calculo impossivel")

#Os parametros y e a não são necessário e servem para transformar na variacao Zipf-Mandelbrot
def distribuicao_zipf(n, y=1, a=0):
    # Gera os ranks e probabilidades de acordo com a lei de Zipf
    ranks = np.arange(1 + a, n + 1 + a)
    probabilities = 1 / (ranks ** y)

    # Normaliza as probabilidades para que somem 1
    probabilities /= np.sum(probabilities)
    return ranks, probabilities

def plot_zipf_distribution(n, y, a=0):
    ranks, probabilities = distribuicao_zipf(n, y, a)

    # Plota a distribuição de Zipf
    plt.plot(ranks, probabilities, color='blue', alpha=0.7)
    plt.xscale('log')
    plt.yscale('log')
    plt.title(f'Distribuição de Zipf {soma} palavras totais e {n} palavras unicas em {nomeLivro} fator y={y}')
    plt.xlabel('Rank')
    plt.ylabel('Probabilidade')

# Mude o caminho para fazer a distribuicao com outros livros
livro = "resultados-pares/saida-casmurro.txt"

#recebe apenas o nome do livro sem o caminho
nomeLivro = livro.split("-")[1]

dicionario = receberDictPar(livro)
listaValores = list(dicionario.values())
listaChaves = list(dicionario.keys())

tamanho = len(listaValores)
soma = sum(listaValores)

# Usando enumerate para obter índices e valores simultaneamente
lista = [i + 1 for i, _ in enumerate(listaValores)]
listaValores = [val / soma for val in listaValores]

# Plota o gráfico de pontos vermelhos
plt.plot(lista, listaValores, 'ro')

n_palavras_diferentes = len(listaValores)
parametro_y = 1

# Chama a função para plotar a distribuição de Zipf no mesmo gráfico
#melhor fator y e aproximadamente 0.6
plot_zipf_distribution(n_palavras_diferentes, 1)

# Exibe o gráfico final
plt.show()
```

# Distribuicao em de palavras em trios

```python
import matplotlib.pyplot as plt
import numpy as np

def receberDictTrio(x=1):
    if x == 1:
        x = input("Arquivo a ser lido: ")
    leitor = open("{}".format(x), "r", encoding="utf-8")
    texto = leitor.read()
    for letra in "(),":
        texto = texto.replace(letra, " ")
    texto = texto.split()

    dicionario = {}
    i = 0
    while i < len(texto):
        # Verifica se a chave é composta por três palavras consecutivas
        chave = f"{texto[i]} {texto[i + 1]} {texto[i + 2]}"
        valor_str = texto[i + 3].strip("',")

        # Avisa que eu fiz algo muito errado pois os valores invadiram as chaves
        if valor_str.isdigit():
            valor = int(valor_str)
            dicionario[chave] = valor
        else:
            print(f"Valor inválido para a chave '{chave}': {valor_str}")

        i += 4  # Avança para o próximo conjunto de palavras e valor

    return dicionario

#Os parametros y e a não são necessário e servem para transformar na variacao Zipf-Mandelbrot
def distribuicao_zipf(n, y=1, a=0):
    # Gera os ranks e probabilidades de acordo com a lei de Zipf
    ranks = np.arange(1, n + 1)
    probabilities = 1 / ((ranks+a) ** y)

    # Normaliza as probabilidades para que somem 1
    probabilities /= np.sum(probabilities)
    return ranks, probabilities

def plot_zipf_distribution(n, y, a=0):
    ranks, probabilities = distribuicao_zipf(n, y, a)

    # Plota a distribuição de Zipf
    plt.plot(ranks, probabilities, color='blue', alpha=0.7)
    plt.xscale('log')
    plt.yscale('log')
    plt.title(f'Distribuição de Zipf {soma} palavras totais e {n} palavras unicas em {nomeLivro} fator y={y}')
    plt.xlabel('Rank')
    plt.ylabel('Probabilidade')

# Mude o caminho para fazer a distribuicao com outros livros
livro = "resultados-trios/saida-casmurro.txt"

#recebe apenas o nome do livro sem o caminho
nomeLivro = livro.split("-")[1]

dicionario = receberDictTrio(livro)
listaValores = list(dicionario.values())
listaChaves = list(dicionario.keys())

tamanho = len(listaValores)
soma = sum(listaValores)

# Usando enumerate para obter índices e valores simultaneamente
lista = [i + 1 for i, _ in enumerate(listaValores)]
listaValores = [val / soma for val in listaValores]

# Plota o gráfico de pontos vermelhos
plt.plot(lista, listaValores, 'ro')

n_palavras_diferentes = len(listaValores)
parametro_y = 1

# Chama a função para plotar a distribuição de Zipf no mesmo gráfico
#aproximadamente y=0.37 e a=0.8 ou a=0
plot_zipf_distribution(n_palavras_diferentes, 1, 0)

# Exibe o gráfico final
plt.show()
```
